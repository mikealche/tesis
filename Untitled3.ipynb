```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu = 400
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

```python
a = tensor(stats.norm.pdf(x, mu, sigma))
b = tensor(stats.norm.pdf(x, mu, sigma))
a = a.reshape(-1,int(math.sqrt(len(a))))
b = a.reshape(-1,int(math.sqrt(len(b))))
```

```python
b = b.transpose(0,1)
ab = (a*b).to(torch.float32)
```
