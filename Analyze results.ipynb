```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
runs = {}
for i in range(1,9):
    runs[i] = pd.read_csv(f'attn_run_seed_{i}.txt')
    runs[i].plot()
```

```python
times_arl_had_higher_acc = [runs[i].max()['Baseline'] < runs[i].max()[' With ARL'] for i in runs]
```

```python
arl_effectiveness = sum(times_arl_had_higher_acc)/ len(runs)
```

```python
print(f'ARL improved the Accuracy in {arl_effectiveness*100}% of the cases')
```
