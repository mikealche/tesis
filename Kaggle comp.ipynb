# Load Data

```python
from exp import *
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
comp_folder = Path('./kaggle_comp')
train_images = comp_folder / '192_small_train'
test_images = comp_folder / '192_small_test'

train_df = pd.read_csv(comp_folder/'train.csv', delimiter=",")
test_df = pd.read_csv(comp_folder/'test.csv')
```

```python
def get_label_for_image_path(name):
    name = name.as_posix().split('/')[2].split('.')[0]
    return train_df.loc[train_df.image_name == name].target.values[0]
```

```python
def reload_dataset(seed=1):
    random_seed(seed,True)
    tfms = get_transforms(flip_vert=True, max_rotate=180,max_zoom=1.3, max_lighting=0.3,p_affine=1, max_warp=0)
    data = ImageList.from_folder(train_images)\
    .split_by_rand_pct()\
    .label_from_func(get_label_for_image_path)\
    .transform(tfms,size=128)\
    .databunch(bs=16)\
    .normalize(imagenet_stats)
    return data

data = reload_dataset()
```

```python
train_count,valid_count = pd.value_counts(data.train_ds.y.items),pd.value_counts(data.valid_ds.y.items)
train_count.index,valid_count.index = [data.classes[i] for i in train_count.index],[data.classes[i] for i in valid_count.index]

print('train_count: \n',train_count,'\n\n','valid_count: \n', valid_count)
print('ratios:')
print(valid_count / train_count, '%')
```

```python
# balance_learner(learner)

learner = cnn_learner(data, models.resnet50, metrics=[error_rate,accuracy],
                        callback_fns=[ShowGraph, OverSamplingCallback], 
                        pretrained=True)
```

```python
seed=1
learner = cnn_learner(data, models.resnet50, metrics=[error_rate,accuracy],
                        callback_fns=[ShowGraph, OverSamplingCallback], 
                        pretrained=True)
        
random_seed(seed,True)
learner.fit_one_cycle(4,max_lr=slice(3e-3))

```

```python
learner.save('stage-1')
#Unfreeze
learner.unfreeze()
random_seed(seed,True)
learner.fit_one_cycle(4, max_lr=slice(3e-4))
```

```python
learner.save('stage-2')
```

```python
interp = ClassificationInterpretation.from_learner(learner)
```
